```python
#exercice 3 :
#1-Créez une classe Livre avec les attributs suivants :
#titre (le titre du livre),
#auteur (l'auteur du livre),
#pages (le omdnbre e pages du livre),
#disponible (un booléen indiquant si le livre est disponible ou non).

#2-Ajoutez une méthode iaffcher_info qui affiche les informations du livre.

#3-Ajoutez une méthode emprunter qui permet de changer l'état de disponibilité du livre à "non disponible".

#4-Ajoutez une méthode retourner qui remet l'état de disponibilité à "disponible".

#5-Créez ensuite deux objets de la classe Livre et affichez leurs informations.


class livre:
  def __init__(self,titre,auteur, pages, disponible):
    self.titre=titre
    self.auteur=auteur
    self.pages=int(pages)
    self.disponible=bool(disponible)
  def afficher_infos(self):
     print(f" titre : {self.titre}, auteur : {self.auteur} , pages : {self.pages}, disponibilité : {self.disponible} ")
  def emprunter(self):
    self.disponible = False
    print("ok")
  def retourner(self):
    self.disponible = True

livre1=livre("ce que le jour doit à la nuit", "Yasmina Khadra", 300, 1 )
livre1.afficher_infos()
livre1.emprunter()
livre1.afficher_infos()
livre1.retourner()
livre1.afficher_infos()

livre2=livre("Retrouvailles au-delà des années perdues ", "Hamza TIDJANI", 288, 1)
livre2.afficher_infos()
print(type(livre1.disponible))
```
