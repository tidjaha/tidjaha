```python
#Exercise 2: List Operations

#  Create a list of your favorite fruits ok.
#  Add a new fruit to the list.
#  Remove a fruit from the list.
#  Sort the list in alphabetical order.
#  Print the list.

l=[]
while True:

  while True:
    a=input("entrer un fruit favoris (le nom d'un fruit) : ")
    try :
     a=float(a)
     print("entrer un nom svp")
    except ValueError:
     break
  l.append(a)
  arret=input("voulez vous entrer un autre fruit ?\ntapez n'importe quelle touche pour continuer OU 'N' pour arreter")
  if arret=="N" or arret=="n":
    break
print("votre liste de fruit favoris est :\n",l)
while True:
    a=input("tapez le nom d'un fruit que vous voulez supprimer de votre liste : ")
    try :
     a=float(a)
     print("entrer un nom svp")
    except ValueError:
     if a in l:
      l.remove(a)
      print(l)
      break
     else :
      print("entrer un nom qui existe dans la liste \n",l)
l.sort()
print("votre liste est :\n",l)
```

```python
#Exercise 4: Set Operations

  #Create two sets of integers.
  #Find the intersection and union of these sets.
  #Print the results.

s1=set([1,2,3,4])
s2=set([1,5,3,6])
setu=s1.union (s2)
print("l'union entre les deux set est :",setu)
setinter= s1.intersection(s2)
print("l'union entre les deux set est :",setinter)
```
